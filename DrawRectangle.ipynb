```python
import turtle 
t = turtle.Turtle()
 
m = 100;
n = 150;
 
t.pencolor ("blue")
t.pensize (8) 

t.fillcolor ("blue")
t.begin_fill()

# vẽ cạnh đầu tiên
t.forward(m) # đi thẳng với m đơn vị
t.left(90)   # quay một góc 90 độ
 
# vẽ cạnh thứ hai
t.forward(n) # đi thẳng với n đơn vị
t.left(90)   # quay một góc 90 độ
 
# vẽ cạnh thứ ba
t.forward(m) # đi thẳng với m đơn vị
t.left(90)   # quay một góc 90 độ
 
# vẽ cạnh thứ tư
t.forward(n) # đi thẳng với n đơn vị
t.left(90)   # quay một góc 90 độ

t.end_fill()

turtle.done()
```
